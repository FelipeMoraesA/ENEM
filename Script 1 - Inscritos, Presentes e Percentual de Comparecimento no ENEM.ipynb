# MICRODADOS ENEM: INSCRITOS VS PRESENTES
#### Autor: Felipe Moraes
O objetivo do presente script é avaliar a evolução da quantidade de inscritos, de presentes e o percentual de comparecimento dos incritos na prova do Enem nos últimos 10 anos (2013-2022). Para tanto, uma mesma rotina é replicada para todos os anos, compilando os dados de cada prova, e ao final é feita uma análise gráfica dos dados extraídos.  
  
**ROTINA:**  
1 - Leitura dos dados de cada ano diretamente do Zip;    
2 - Filtragem das colunas de interesse;  
3 - Mensuração dos incritos, dos presentes e do percentual de comparecimento;  
4 - Inclusão das informações no Data.Frame próprio;  
5 - Visualização dos dados extraídos para aquele ano.   
  
**NOTAS:**
> O conceito de **presentes** consiste nos inscritos que realizaram *100%* da prova. Isto é, compareceram os dois dias de prova e não foram eliminados em nunhum dos dias.  
  
> Por **percentual de comparecimento dos inscritos** se entende a quantidade de pessoas presentes na prova em relação ao total de inscritos em cada ano.

```python
# Pacotes Utilizados:
import pandas as pd
from matplotlib import pyplot as plt
import zipfile
import seaborn as sns
```

### Dados de 2022

```python
# leitura dos dados 2022:
with zipfile.ZipFile('Dados/Dados_Brutos/microdados_enem_2022.zip') as zip_file:
    with zip_file.open('DADOS/MICRODADOS_ENEM_2022.csv') as csv_file:
        dadosbrutos = pd.read_csv(csv_file, sep=';', encoding='ISO-8859-1')

# Filtragem das colunas de interesse:
enem2022 = dadosbrutos.filter(['NU_ANO', 'NU_INSCRICAO','TP_PRESENCA_CN','TP_PRESENCA_CH','TP_PRESENCA_LC','TP_PRESENCA_MT'])\
    .reset_index(drop=True)

# Mensuração dos inscritos e dos presentes
inscritos = enem2022['NU_INSCRICAO'].nunique()
presentes = (enem2022.loc[:, 'TP_PRESENCA_CN':'TP_PRESENCA_MT'] == 1).all(axis=1).sum()
per_compare = ((presentes/inscritos)*100).round(2)

# Criação do Data.Frame de compilação:
dados = {'Ano':[], 'Inscritos':[], 'Presentes':[], '%_de_Comparecimento':[]}
dados = pd.DataFrame(dados)
n=0

# Inserção dos dados no Data.Frame de compilação
dados.loc[n] = ['2022', inscritos, presentes, per_compare]

# Visualização dos dados:
display(enem2022)
display(dados)
```

### Elaboração do Gráfico

```python
# Gráfico 1: Inscritos vs Presentes
fig, ax = plt.subplots(figsize=(12, 7))

ax.plot(dados['Ano'], dados['Inscritos'], label = 'Inscritos', color = 'blue', ls = '-')
ax.plot(dados['Ano'], dados['Presentes'], label = 'Presentes', color = 'green', ls = '-')

# Eixo X
ax.set_xlim([2012.6, 2022.3])
plt.xticks([2013,2014,2015,2016,2017,2018,2019,2020,2021,2022])

# Eixo Y
ax.set_ylabel('Candidatos em Milhões', size=12)

ax.set_ylim(1e6, 10e6)
labels = [f'{val/1e6:.1f}' for val in ax.get_yticks()]
ax.set_yticklabels(labels)

# Título
ax.set_title('ENEM: Inscritos e Presentes (2013-2022)', size=14)

# Legenda
ax.legend(bbox_to_anchor=(0.04,1), loc='upper left')

# Nota de rodapé
ax.text(0.0, -0.115, 'Elaboração: Felipe Moraes \nFonte: Microdados do ENEM (INEP)',
        transform=ax.transAxes, ha='left')

sns.despine()

plt.show()
```
